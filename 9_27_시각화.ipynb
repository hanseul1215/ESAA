```python
# 엔스콤 데이터 집합
import seaborn as sns
anscombe = sns.load_dataset("anscombe")
print(anscombe)
print(type(anscombe))
```

```python
# matplotlib 라이브러리
%matplotlib notebook
import matplotlib.pyplot as plt
dataset_1 = anscombe[anscombe['dataset']=='I']
plt.plot(dataset_1['x'], dataset_1['y'])
plt.plot(dataset_1['x'], dataset_1['y'], 'o') # 점으로
```

```python
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
```

```python
#################1. Matplotlib 라이브러리로 그래프 그리기
# 1.기본틀 만들기
fig = plt.figure()
# 2. 그래프 격자 만들기 / add_subplot(행크기, 열크기, index)
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
# 3. 그래프 그리기
axes1.plot(dataset_1['x'], dataset_1['y'],'o')
axes2.plot(dataset_2['x'], dataset_2['y'],'o')
axes3.plot(dataset_3['x'], dataset_3['y'],'o')
axes4.plot(dataset_4['x'], dataset_4['y'],'o')

#제목 추가 / set_title
axes1.set_title("dataset_1")
axes3.set_title("dataset_3")
fig.suptitle("Anscombe Data")
# 그래프 레이아웃 / tight_layout
fig.tight_layout()
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
print(tips.describe())
```

```python
# 히스토그램 : 데이터의 분포와 빈도를 살피는 용도
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)

axes1.hist(tips["total_bill"], bins=10) # bins x축 간격 10으로 조정
axes1.set_title('Histo')
axes1.set_xlabel('Total bill')
axes1.set_ylabel('Frequency')
```

```python
# 산점도 : 변수 2개를 사용하는 이변량 그래프
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title("scatter")
axes1.set_xlabel('Total bill')
axes1.set_ylabel('tip')
```

```python
tips[tips['sex']=='Female']['tip']
```

```python
# 박스 그래프 : 이산형 변수와 연속형 변수를 함께 사용하는 그래프
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
              tips[tips['sex']=='Male']['tip']],
             labels=['female','male'])

axes1.set_xlabel('sex')
axes1.set_ylabel('tip')
```

```python
# 다변량 그래프 그리기
## 산점도
# 문자열은 산점도 그래프 색상 지정 못해
def recode_sex(x):
    if x == "Female":
        return 0
    else:
        return 1
tips['sex_color']=tips["sex"].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

axes1.scatter(tips['total_bill'],
              tips['tip'], 
              s=tips['size']*10, 
              c=tips['sex_color'], alpha=0.5) #s : 크기, c: 색상 / alpha : 투명도
axes1.set_title("scatter")
axes1.set_xlabel('Total bill')
axes1.set_ylabel('tip')
```

```python
# seaborn 라이브러리 사용하기

# 단변량 -히스토그램 / subplots, distplot 사용
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True) # rug : 양탄자 그래프, 그래프의 축에 동일한 길이의 직선을 붙여 밀집 정도를 표현
ax.set_title('total bill histo') 

# 밀집도 제외 distplot kde = False / 히스토그램 제외 hist=False
```

```python
# count 그래프, 이산값을 나타낸 그래프
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_xlabel("day")
ax.set_ylabel("freqency")
```

```python
# 이변량 - 산점도 / regplot : 산점도와 회귀선 같이 그려
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips) # fit_reg=False
ax.set_xlabel('total_bill')
ax.set_ylabel('tip')
```

```python
# 산점도와 히스토 한번에 / jointplot
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='total bill', ylabel='tip')
joint.fig.suptitle('joint')
```

```python
# 산점도 -> 육각그래프, 데이터 구분 쉬어, 색의 진하기로 밀도 표현
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind ='hex') # kind ='hex' :육각형
```

```python
# 이차원 밀집도 그리기 / kdeplot
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'], data2 = tips['tip'], shade=True) #shade = True 음영효과
ax.set_xlabel('Total bill')
ax.set_ylabel('tip')
```

```python
# 바 그래프 그리기 : 지정한 변수의 평균을 계산하여 그릴 수 있다.
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('bar plot')
ax.set_xlabel('day')
ax.set_ylabel('average total bill')
```

```python
# 박스 그래프
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title("box plot")
ax.set_xlabel("day")
ax.set_ylabel("total bill")
```

```python
# 바이올링 그래프 / violinplot
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title("violin plot")
ax.set_xlabel("day")
ax.set_ylabel("total bill")
```

```python
# 관계 그래프 / pairplot
#tips[["total_bill"],["tip"],["size"]]
Tips=tips.loc[:,["total_bill", "tip", "size"]]
fig = sns.pairplot(Tips)
```

```python
# map_upper / map_lower / map_diag
pair_grid = sns.PairGrid(Tips)
pair_grid = pair_grid.map_upper(sns.regplot) 
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()
```

```python
# 다변량 그래프 - seaborn
# 바이올린
ax = plt.subplots()
ax = sns.violinplot(x =  'time', y = 'total_bill', hue = 'sex', data = tips, split= True) # hue 인자로 색상추가
```

```python
# seaborn -> 그래프 스타일 설정하기 /set_style
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
fig = plt.figure()

seaborn_styles =['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx +1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
